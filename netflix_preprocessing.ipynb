# Preprocessing Netflix Dataset
This notebook will take the `netflix.csv` dataset, preprocess it by handling missing values, converting temporal variables, and extracting meaningful features out of existing ones. Then it saves the cleaned data for further use cases.

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Load the dataset
df = pd.read_csv('netflix.csv')
print(f"Original dataset shape: {df.shape}")
df.head()
```

## 1. Handling Missing Values
We can fill unknown directors, casts, and countries with 'Unknown' rather than dropping rows, as this data is valuable. The missing `date_added` and `rating` are very few so we will drop them.

```python
# Check for missing values
print("Missing values before:")
missing = df.isnull().sum()
print(missing[missing > 0])

# Filling categorical features with large amount of NaNs
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

# For 'date_added', 'rating', and 'duration', dropping the few rows with NaN
df.dropna(subset=['date_added', 'rating', 'duration'], inplace=True)

print("\nMissing values after processing:", df.isnull().sum().sum())
```

## 2. Formatting Variables (Dates)
We convert `date_added` to a datetime object, and extract the `year_added` and `month_added`.

```python
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'])

# Extract useful temporal features
df['year_added'] = df['date_added'].dt.year.astype(int)
df['month_added'] = df['date_added'].dt.month.astype(int)

df[['date_added', 'year_added', 'month_added']].head()
```

## 3. Formatting Variables (Duration)
Separate the duration column into a numeric format `duration_num`.

```python
# Extract numeric components
df['duration_num'] = df['duration'].str.extract(r'(\d+)').astype(int)
df[['type', 'duration', 'duration_num']].head()
```

## 4. Final Data Overview and Save

```python
print(f"Cleaned dataset shape: {df.shape}")
output_path = 'netflix_cleaned.csv'
df.to_csv(output_path, index=False)
print(f"Preprocessed dataset saved successfully as {output_path}!")
df.info()
```
